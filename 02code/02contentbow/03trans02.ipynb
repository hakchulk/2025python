```python
import pandas as pd
from ast import literal_eval
import seaborn as sns
import numpy as np


movies_metadata = pd.read_csv("data/movies_metadata.csv")
links_small  = pd.read_csv("data/links_small.csv")
movies_keywords  = pd.read_csv("data/keywords.csv")

links_small = links_small[links_small['tmdbId'].notnull()]["tmdbId"].astype('int')
movies_metadata_small = movies_metadata[movies_metadata['id'].isin(links_small.astype('str'))]

movies_metadata_small.shape
```

```python
movies_keywords['id'] = movies_keywords['id'].astype('str') # int -> str
```

```python
movies_metadata_small = movies_metadata_small.merge(movies_keywords,on='id')
movies_metadata_small.shape
```

```python
movies_metadata_small.head()
```

```python
movies = movies_metadata_small[['id','title','genres','keywords', 'overview','popularity', 'release_date']]
```

```python
movies.info()
```

```python
movies['str_genres'] = movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: sorted(i['name'] for i in x) if isinstance(x, list) else []) \
    .apply(lambda x: ', '.join(x) if len(x) > 0 else None)
```

```python
movies.head()
```

```python
movies['str_keywords'] = movies['keywords'].fillna('[]').apply(literal_eval).apply(lambda x: sorted(i['name'] for i in x) if isinstance(x, list) else []) \
    .apply(lambda x: ', '.join(x) if len(x) > 0 else None)
```

```python
movies.head()
```

```python
def get_genres_keysord_overview(row):
    return f"""
    genres:
    {row['str_genres']}

    keywords:
    {row['str_keywords']}

    overview:
    {row['overview']}

    """
```

```python
movies['genres_keysord_overview'] = movies.apply(lambda row: get_genres_keysord_overview(row), axis=1)
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['year'] = movies['release_date'].dt.year

movies['popularity'] = movies['popularity'].astype(float)
movies['popularity_log'] = np.log(movies['popularity'])

movies = movies.dropna().reset_index(drop=True)
```

```python
from transformers import BertTokenizer, BertModel
import torch # transformers ÏÇ¨Ïö©ÏùÑ ÏúÑÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨
model_name = 'bert-base-uncased'
```

```python
# ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†Ä(Tokenizer) : ÌÖçÏä§Ìä∏Î•º Î™®Îç∏Ïù¥ Ï≤òÎ¶¨Ìï† Ïàò ÏûàÎäî Ïà´Ïûê(ID)Î°ú Î∞îÍæ∏Îäî Îã§Î¶¨ Ïó≠Ìï†
import os
from transformers import BertTokenizer

model_name = 'bert-base-uncased'
save_path = './my_tokenizer'  # Ï†ÄÏû•Ìï† ÎîîÎ†âÌÜ†Î¶¨ Í≤ΩÎ°ú

# 1. Í≤ΩÎ°úÏóê ÌååÏùºÏù¥ ÏûàÎäîÏßÄ ÌôïÏù∏
if os.path.exists(save_path) and os.listdir(save_path):
    # ÌååÏùºÏù¥ ÏûàÏúºÎ©¥ Î°úÏª¨ÏóêÏÑú Î°úÎìú
    print("Î°úÏª¨ÏóêÏÑú ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†ÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§.")
    tokenizer = BertTokenizer.from_pretrained(save_path)
else:
    # ÌååÏùºÏù¥ ÏóÜÏúºÎ©¥ Ïò®ÎùºÏù∏ÏóêÏÑú Îã§Ïö¥Î°úÎìú ÌõÑ Ï†ÄÏû•
    print("ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†ÄÎ•º ÏÉàÎ°ú Îã§Ïö¥Î°úÎìúÌïòÍ≥† Ï†ÄÏû•Ìï©ÎãàÎã§.")
    tokenizer = BertTokenizer.from_pretrained(model_name)
    
    # ÎîîÎ†âÌÜ†Î¶¨Í∞Ä ÏóÜÏúºÎ©¥ ÏÉùÏÑ± ÌõÑ Ï†ÄÏû•
    os.makedirs(save_path, exist_ok=True)
    tokenizer.save_pretrained(save_path)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"ÌòÑÏû¨ ÏÇ¨Ïö© Ï§ëÏù∏ Ïû•Ïπò: {device}")
```

```python
model = BertModel.from_pretrained(model_name)
model.to(device)
model.eval()  # ÌèâÍ∞Ä Î™®ÎìúÎ°ú ÏÑ§Ï†ï (ÌïôÏäµÏù¥ ÏïÑÎãå Ï∂îÎ°†Ïö©)
```

```python
# Î≤°ÌÑ∞(ÏûÑÎ≤†Îî©)Î•º Ï∂îÏ∂ú
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True) # PyTorch ÌÖêÏÑúÎ°ú Î∞òÌôò
    # ÏûÖÎ†•ÏùÑ Î™®Îç∏Ïù¥ ÏûàÎäî Ïû•Ïπò(GPU)Î°ú Ïù¥Îèô
    # inputs = {k: v.to(device) for k, v in inputs.items()}
    inputs = inputs.to(device)
    with torch.no_grad():  # Í∑∏ÎûòÎîîÏñ∏Ìä∏ Í≥ÑÏÇ∞ ÎπÑÌôúÏÑ±Ìôî
        outputs = model(**inputs) ## Î™®Îç∏Ïóê ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÎã¨ÌïòÏó¨ Ï∂úÎ†• ÌöçÎìù.  **Îäî "ÎîïÏÖîÎÑàÎ¶¨ Ïñ∏Ìå®ÌÇπ(Dictionary Unpacking)" Ïó∞ÏÇ∞Ïûê
    return outputs.last_hidden_state[:,0,:].squeeze().cpu().numpy()  # [Î∞∞Ïπò ÌÅ¨Í∏∞, ÏãúÌÄÄÏä§ Í∏∏Ïù¥, ÌûàÎì† ÌÅ¨Í∏∞]ÏóêÏÑú Ï≤´ Î≤àÏß∏ ÌÜ†ÌÅ∞([CLS])Ïùò ÏûÑÎ≤†Îî© Î∞òÌôò

```

```python
from tqdm import tqdm
tqdm.pandas()
movies['embedding'] = movies['genres_keysord_overview'].progress_apply(get_bert_embedding)
# cpu : 8it/s
# gpu : 40it/s
```

```python
# ÌòÑÏû¨ ÏÇ¨Ïö© Ï§ëÏù∏ GPUÏùò Î©îÎ™®Î¶¨ Ï†êÏú†Ïú® ÌôïÏù∏
print(f"Allocated: {torch.cuda.memory_allocated(0) / 1024**2:.1f}MB")
print(f"Cached: {torch.cuda.memory_reserved(0) / 1024**2:.1f}MB")
```

```python
all_embeddings = movies['embedding'].values.tolist()
```

```python
import numpy as np

all_embeddings = np.array(all_embeddings)
all_embeddings.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_of_overview = cosine_similarity(all_embeddings,all_embeddings)
similarity_of_overview.shape
```

```python
sorted_similarity_of_overview = similarity_of_overview.argsort()[:,::-1]
```

```python
def recom(title_name,top_k=30):
    movie_of_title = movies[movies['title'] == title_name]
    print(f'{title_name}Ïùò Ïû•Î•¥ : {movie_of_title['str_genres'].values[0]}')

    movies_index_of_title = movie_of_title.index.values[0]

    similar_indexes = sorted_similarity_of_overview[movies_index_of_title,:top_k*2]

    similar_indexes = similar_indexes.reshape(-1)

    similar_indexes = similar_indexes[similar_indexes != movies_index_of_title]

    return movies.iloc[similar_indexes].sort_values(by=['popularity','year'],ascending=False)[:5]
```

```python
recom_movies = recom('Mulan')
recom_movies[['title','year','str_genres','popularity','str_keywords','overview']]
```
