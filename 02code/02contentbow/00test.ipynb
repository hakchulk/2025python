```python
import torch

# 1. GPU 가속 가능 여부
is_cuda = torch.cuda.is_available()
print(f"CUDA 사용 가능: {is_cuda}")

if is_cuda:
    # 2. 현재 설치된 PyTorch가 인식하는 CUDA 버전
    print(f"PyTorch의 CUDA 버전: {torch.version.cuda}")
    # 3. GPU 상세 정보
    print(f"GPU 이름: {torch.cuda.get_device_name(0)}")
    print(f"현재 할당된 메모리: {torch.cuda.memory_allocated(0) / 1024**2:.2f} MB")
    # 현재 사용 중인 GPU의 메모리 점유율 확인
    print(f"Allocated: {torch.cuda.memory_allocated(0) / 1024**2:.1f}MB")
    print(f"Cached: {torch.cuda.memory_reserved(0) / 1024**2:.1f}MB")
```
