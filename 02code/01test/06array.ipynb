```python
arr20 = np.arange(1, 10, 1)
print(arr20)
arr21 = arr20.reshape(3, 3)
print(arr21)

# [0:2, 0:2]
arr31 = arr21[0:2, 0:2]
print(arr31, arr31.shape)

# [1:3, 0:3] = [1:3, :]
arr32 = arr21[1:3, 0:3]
print(arr32, arr32.shape)

# [:2, :1]
# [:2, 0]
arr33 = arr21[:2, :1]
print(arr33, arr33.shape)
arr34 = arr21[:2, 0]
print(arr34, arr34.shape)
```

### 동물별 평균체온과 최대 체중 구하기

```python
import numpy as np
# 체온, 심박수, 체중(kg)
data = np.array([    # axis=0 행
    [37.5, 60, 10],  # axis=1 열
    [39.1, 120, 3.5], 
    [37.2, 28, 460] ])

ani = np.array(['dog', 'cat', 'horse'])
avg_values = data.mean(axis=0)
weight = data[:, 2]
max_weight = weight.max()
max_ani = ani[data[:,2].argmax()]
print("동물별 평균:", avg_values)
print("가장 무거운 동물:", max_ani, f'{max_weight} Kg')
```

### 곤충 [날개수, 다리수, 몸길이(cm)]

```python
insect = np.array(["나비", "잠자리", "개미", "거미"])
data = np.array([
    [4, 6, 6],
    [4, 6, 8],
    [0, 6, 1],
    [0, 8, 2]]
)
```

```python
# 다리가 6개인 곤충만
print(data[:,1] == 6)
legs = data[data[:,1] == 6]
# legs = data[[True,  True,  True, False]]
print(legs)

# 날개가 없는 곤충
no_wings = insect[data[:,0] == 0]
print(no_wings)

mean_len = data[:,2].mean()
print("평균 몸길이:", mean_len)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nation = ["kor", "usa", "japan", "france"]
men = [175.5, 179.9, 168.1, 178.6]
women = [163.2, 165.9, 158.1, 168.1]
plt.figure(figsize=(5,3))
plt.bar(nation, men)
plt.bar(nation, women)
plt.show()
```

```python
nation = ["kor", "usa", "japan", "france"]
men = [175.5, 179.9, 168.1, 178.6]
women = [163.2, 165.9, 158.1, 168.1]
ind = np.arange(len(nation))
print(ind,type(ind))
w = 0.1

plt.figure(figsize=(5,3))
plt.bar(ind, men, width=w, color = "orange", label = "mem")
plt.bar(ind, women, bottom=men, width=w, color = "b", label = "women")
plt.xticks(ind, nation)
plt.legend()
plt.show()
```
