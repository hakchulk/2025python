```python
import matplotlib.pyplot as plt
# 대한민국 명목 GDP (Billion USD)
year = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
gdp = [1484.0, 1466.0, 1500.0, 1623.0, 1725.0, 1651.0, 1644.0, 1811.0, 1673.3, 1712.8, 1760.9]
```

```python
plt.plot(year, gdp)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(year, gdp, "--o")
plt.title("GDP of Korea")
plt.xlabel("Year")
plt.ylabel("GDP in Billion USD")
```

```python
year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
seoul = [12.1,12,12.2,12.5,13.4,13.6,13.6,13,12.9,13.5,13.2,13.7,13.2]
jeju = [15.6,15.6,15.7,16.5,16.2,16.7,17,16.8,16.6,16.8,16.7,17.5,17]
plt.figure(figsize=(5,5))
plt.plot(year, seoul, label="Seoul")
plt.plot(year, jeju, label="Jeju")
plt.title("Average Temperature in Korea")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python
import numpy as np

width = 0.3
x = np.arange(len(year))
plt.figure(figsize=(5,6))
plt.bar(x - width/2, jeju, width=width, label="Jeju")
plt.bar(x + width/2, seoul, width=width, label="Seoul")
plt.title("Average Temperature in Korea")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python
import csv
tcount = 0 # 전체
count = 0  # 유효
htemp = [] # 최고기온 리스트

with open('data/seouldata2.csv', encoding='utf-8') as tfile:
    data = csv.reader(tfile)
    next(data)  # 헤더 스킵
    for row in data:
        tcount += 1
        if row[-1] != '' :
            count += 1
            htemp.append(float(row[-1])) # 최고기온 데이터 추가

print("전체 데이터 건수:", tcount)
print("유효 데이터 건수:", count)
print("결측치 데이터 건수:", tcount - count)
# plt.plot(htemp)
plt.boxplot(htemp)
#plt.hist(htemp, bins=100)


```
