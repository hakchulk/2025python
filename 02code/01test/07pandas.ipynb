```python
df = pd.DataFrame({
    'name': ['Jessica', 'Liam', 'Sophia', 'Ryan', 'Alex'],
    'test': [45, 30, 40, 37, 48],
    'assign1': [20, 17, 22, 18, 24],
    'assign2': [19, 14, 18, 15, 25]
})
df
```

```python
df.rename(columns={'test' : 'exam'}, inplace='True')
```

```python
df['sum'] = df[['exam', 'assign1', 'assign2']].sum(axis=1)
df
```

```python
df1 = df[['name','sum']]
df1
```

```python
df3 = df[df['sum'] >= 80][['sum', 'name']]
```

```python
df5 = df3.sort_values(by='sum' , ascending=False).reset_index(drop=False)
df5
```

```python
# df['grade'] = np.where(
#     df['sum'] >= 90, 'A', 
#     np.where(df['sum'] >= 80, 'B','C')
#     )

bins = [0, 70, 80, 90, float('inf')]
labels = ['D', 'C', 'B', 'A']

# right=False는 '미만' 기준으로 경계선을 설정함 (80점부터 B, 90점부터 A)
df['grade'] = pd.cut(df['sum'], bins=bins, labels=labels, right=False)
df
```

```python
df['result'] = np.where(df['sum'] >= 80, 'Pass', 'Fail')
df
```

```python
df_count = df['result'].value_counts()
df_count.shape
```

```python

df_c = df_count.to_frame()
df_c
```

```python
df_c.reset_index(drop=True)
df_c
```
