```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 로드 (파일 경로를 실제 csv 파일명으로 수정하세요)
# 예: data.csv 파일에 헤더가 없는 경우를 가정하여 컬럼명을 지정합니다.

df = pd.read_csv('data/seouldata2.csv')

# 2. 데이터 전처리
# 컬럼명에 공백이 있을 수 있으므로 정리해주고, 날짜 데이터를 변환합니다.
df.columns = [col.strip() for col in df.columns]  # 컬럼명 양끝 공백 제거
df['날짜'] = pd.to_datetime(df['날짜'].str.strip()) # 날짜 값의 공백 제거 및 변환
df['연도'] = df['날짜'].dt.year

# 3. 한글 깨짐 방지 설정 (Windows 환경)
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

# 4. 시각화: 연도별 최고기온(℃) 박스 플롯
plt.figure(figsize=(15, 7))
sns.boxplot(x='연도', y='최고기온(℃)', data=df, palette='Oranges')

# 그래프 상세 설정
plt.title('연도별 최고기온 분포 (4100일 데이터)', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('최고기온(℃)', fontsize=12)
plt.xticks(rotation=45) # 연도가 많을 경우 글자가 겹치지 않게 회전
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
