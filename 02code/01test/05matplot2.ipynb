```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,4,9,16]
y1 = y[::-1]
plt.figure(figsize=(4,3))
plt.title("Sample bar")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, '--o', color='#ff0000', linestyle=':', marker='d')
plt.plot(x,y1, '--o', color='blue', linestyle=':', marker='d')
plt.xlim(0,5)
plt.ylim(0,20)
# plt.bar(x,y )
plt.show()
```
