```python
import pandas as pd
df1 = pd.DataFrame({
    'name': ['Jessica', 'Liam', 'Sophia', 'Ryan', 'Alex'],
    'gender': ['F', 'M', 'F', 'M', 'F'],
    'age': [21, 23, 20, 22, 20]
})
df2 = pd.DataFrame({
    'name': ['Sophia', 'Liam', 'Ryan', 'Alex', 'Jessica'],
    'department': ['computer science', 'mathematics', 'law',
                   'computer science', 'law']
})

df4 = pd.DataFrame({
    'department': ['computer science', 'mathematics', 'law'],
    'head of department': ['Steve', 'Emma', 'Carly']
})
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2) # = pd.merge(df1, df2, on='name')
df3
```

```python
df5 = pd.merge(df3, df4)
df5
```

```python
df5 = pd.merge(df3, df4, on ='department')
df5
```

```python
df5.sort_values('age', ascending=False)
```
