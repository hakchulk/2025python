```python
import pandas as pd

data = {
    "class": ["A", "A", "B", "B"],
    "name": ["철수", "영희", "민수", "지수"],
    "score": [80, 90, 70, 85]
}
df = pd.DataFrame(data)
print(data)
df
```

```python
gr = df.groupby('class')

for key, value in gr['score']:
    print(key)
    print(value)
    print(type(value))
```

```python
print(df.groupby('class')['score'].mean())
```

```python
data1 = {
    '학급': ['A', 'A', 'B', 'B', 'C', 'C'],
    '이름': ['철수', '영희', '민수', '수진', '지훈', '혜진'],
    '수학': [90, 80, 70, 85, 60, 95],
    '영어': [85, 88, 75, 70, 95, 90]
}
df1 = pd.DataFrame(data1)
df1_mean = df1.groupby('학급')[ ['수학', '영어'] ].mean().reset_index()
print(df1_mean)
print(df1_mean.shape)
```

```python
df2_mean = df1.groupby('학급')[ ['수학', '영어'] ].mean()
print(df2_mean)
print(df2_mean.shape)
```

```python
gr = df1.groupby('학급')[ '수학' ].mean()
print(gr)
print(gr.shape)
```

```python
gr = df1.groupby('학급')[ '수학' ].agg(['mean', 'max', 'count'])
gr
```
