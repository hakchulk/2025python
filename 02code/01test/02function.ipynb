```python
import random
def rand_add():
    a = random.randint(1, 100)
    b = random.randint(1, 100)
    answer = int(input(f"What is {a} + {b}? "))
    if answer == a + b:
        print("Correct!")
    else:
        print(f"Incorrect! The correct answer is {a + b}.")
    

rand_add()
    
```

### List Comprehention 리스트 내포
- [표현식 for 변수 in 반복]
- [표현식 for 변수 in 반복 if 조건]
- [참 if 조건 else 거짓]

```python
a = [i for i in range(1,10,2)]
print(a)
```

```python
b = [i + 1 for i in range(1,10) if i % 3 == 0]
print(b)
```

```python
c = [x for x in [-2, -1, 0,2,5] if x > 0]
c
```

```python
score = 15
result = "합격" if score > 60 else "불합격"
result
```

```python
num = [95, 70, 40, 85, 60]
d = ['합격:' + str(s) if s > 60 else '불합격:' + str(s) for s in num]
d
```

```python
def sum(n:int) -> int:
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(sum(9))
```

```python
def sum1(n:int) -> int:
    total = 0
    a = 1
    while a <= n:
        total += n
        n -= 1
    return total

sum1(9)
```
